```python
# masukan library yang dibutuhkan
import pandas as pd
```

```python
# masukan  function yang dibutuhkan
def count(labelname1, val1, labelname2, val2): #menghitung banyaknya data berlabel 'labelname1' = val1 ketika data berlabel 'labelname2' = val
    i = 0
    for j in range(len(datatrain)):
        if(datatrain[labelname1][j] == val1) & (datatrain[labelname2][j] == val2):
            i = i + 1
    return i

def naivebayes(labelname1, val1, labelname2, val2, income_type_amount): #menghitung peluang menggunakan rumus naive bayes
    result = count(labelname1, val1, labelname2, val2) / income_type_amount
    return result
```

```python
# View data train
datatrain = pd.read_excel('TrainsetTugas1ML.xlsx')
datatrain
```

```python
# View data test
datatest = pd.read_excel('TestsetTugas1ML.xlsx')
datatest
```

```python
#langkah 1 : menentukan peluang income = '>50K' dan income = '<=50K'
up_income = 0 #menghitung banyaknya income >50K
down_income = 0 #menghitung banyaknya income <=50K
for a in range(len(datatrain)):
    if(datatrain['income'][a]=='>50K'):
        up_income = up_income + 1
    else:
        down_income = down_income + 1
Pup_income = up_income/len(datatrain) #peluang income = '>50K'
Pdown_income = down_income/len(datatrain) #peluang income = '<=50K'
```

```python
# Langkah 2 : menentukan peluang dari masing-masing atribut
datatest_income = []
for i in range(len(datatest)):
    Page_up = naivebayes('age', datatest['age'][i], 'income', '>50K', up_income)
    Page_down = naivebayes('age', datatest['age'][i], 'income', '<=50K', down_income)
    
    Pworkclass_up = naivebayes('workclass', datatest['workclass'][i], 'income', '>50K', up_income)
    Pworkclass_down = naivebayes('workclass', datatest['workclass'][i], 'income', '<=50K', down_income)
    
    Peducation_up = naivebayes('education', datatest['education'][i], 'income', '>50K', up_income)
    Peducation_down = naivebayes('education', datatest['education'][i], 'income', '<=50K', down_income)
    
    Pmaritalstatus_up = naivebayes('marital-status', datatest['marital-status'][i], 'income', '>50K', up_income)
    Pmaritalstatus_down = naivebayes('marital-status', datatest['marital-status'][i], 'income', '<=50K', down_income)
    
    Poccupation_up = naivebayes('occupation', datatest['occupation'][i], 'income', '>50K', up_income)
    Poccupation_down = naivebayes('occupation', datatest['occupation'][i], 'income', '<=50K', down_income)
    
    Prelationship_up = naivebayes('relationship', datatest['relationship'][i], 'income', '>50K', up_income)
    Prelationship_down = naivebayes('relationship', datatest['relationship'][i], 'income', '<=50K', down_income)
    
    Phoursperweek_up = naivebayes('hours-per-week', datatest['hours-per-week'][i], 'income', '>50K', up_income)
    Phoursperweek_down = naivebayes('hours-per-week', datatest['hours-per-week'][i], 'income', '<=50K', down_income)
    
    Pup = Page_up*Pworkclass_up*Peducation_up*Pmaritalstatus_up*Poccupation_up*Prelationship_up*Phoursperweek_up*Pup_income
    Pdown = Page_down*Pworkclass_down*Peducation_down*Pmaritalstatus_down*Poccupation_down*Prelationship_down*Phoursperweek_down*Pdown_income
    
    # Langkah 3: hasil prediksi dimasukkan ke array datatest_income
    if(Pup > Pdown):
        datatest_income.append('>50K')
    else:
        datatest_income.append('<=50K')
```

```python
# Langkah 4: masukkan hasil prediksi income ke file datatest
TebakanTugas1 = pd.DataFrame({'income': datatest_income}, index=datatest['id'])
TebakanTugas1.to_excel("TebakanTugas1.xlsx")
```
